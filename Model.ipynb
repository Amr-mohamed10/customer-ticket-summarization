```python
# Step 1: Import necessary libraries
import pandas as pd
import io # We might need this if reading from a string later, good to have

print("Pandas imported successfully.")
```

```python
# Step 2: Load the data from the text file

# Define the path to your data file (assuming the notebook is in the 'Vois' folder)
# If your notebook is *outside* the Vois folder, adjust the path accordingly.
# e.g., file_path = 'Vois/Ticket Data.txt'
file_path = 'Ticket Data.txt' # Use this if notebook is inside Vois folder

try:
    # Read the comma-separated text file into a pandas DataFrame
    df_tickets = pd.read_csv(file_path)
    print("Ticket data loaded successfully!")
    print(f"Total tickets loaded: {len(df_tickets)}")
    print("\nFirst 5 rows of the data:")
    # Display the first 5 rows using markdown for better notebook formatting
    print(df_tickets.head().to_markdown(index=False))
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    print("Please ensure 'Ticket Data.txt' is in the same directory as the notebook, or adjust the file_path variable.")
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
```

```python
# Step 3: Filter Data by SERVICE_CATEGORY

# Check if df_tickets was loaded successfully before proceeding
if 'df_tickets' in locals():
    # Define the list of service categories to keep
    required_categories = ['HDW', 'NET', 'KAI', 'KAV', 'GIGA', 'VOD', 'KAD'] # [cite: 8]

    # Create the filtered DataFrame
    df_filtered = df_tickets[df_tickets['SERVICE_CATEGORY'].isin(required_categories)].copy()

    print(f"Original number of tickets: {len(df_tickets)}")
    print(f"Number of tickets after filtering by SERVICE_CATEGORY: {len(df_filtered)}")

    # Verify the remaining categories
    print("\nUnique SERVICE_CATEGORY values remaining:")
    print(df_filtered['SERVICE_CATEGORY'].unique())

    print("\nFirst 5 rows of filtered data:")
    print(df_filtered.head().to_markdown(index=False))
else:
    print("Error: df_tickets DataFrame not found. Please ensure Step 2 ran successfully.")
```

```python
# Get all column names
column_names = df_filtered.columns.tolist()
print(column_names)
```

```python
# Step 4: Select a Specific Customer

# Check if df_filtered exists
if 'df_filtered' in locals():
    # --- Simulation of customer selection ---
    # In the notebook, we manually set the customer number.
    # In Streamlit, this would come from st.selectbox or similar.
    selected_customer = '123'
    # --- End Simulation ---

    # Filter the data further for the selected customer
    df_customer = df_filtered[df_filtered['CUSTOMER_NUMBER'].astype(str) == selected_customer].copy()
    # Using .astype(str) in case customer numbers are read as integers

    print(f"\nFiltering for CUSTOMER_NUMBER: {selected_customer}")
    print(f"Number of tickets for customer {selected_customer}: {len(df_customer)}")

    if not df_customer.empty:
        print(f"\nFirst 5 rows of data for customer {selected_customer}:")
        print(df_customer.head().to_markdown(index=False))
    else:
        print(f"No tickets found for customer {selected_customer} within the filtered categories.")
else:
    print("Error: df_filtered DataFrame not found. Please ensure Step 3 ran successfully.")
```

```python
# Step 5: Map Service Category to Product Category

# Check if df_customer exists and is not empty
if 'df_customer' in locals() and not df_customer.empty:
    # Define the mapping dictionary [cite: 16]
    category_to_product = {
        'KAI': 'Broadband',
        'NET': 'Broadband',
        'KAV': 'Voice',
        'KAD': 'TV',
        'GIGA': 'GIGA',
        'VOD': 'VOD',
        # Note: 'HDW' is filtered in but not mapped here per the doc.
        # We need to decide how to handle 'HDW'. Let's map it to 'Hardware' for now.
        'HDW': 'Hardware'
    }

    # Apply the mapping to create the new column
    df_customer['PRODUCT_CATEGORY'] = df_customer['SERVICE_CATEGORY'].map(category_to_product)

    print("\nAdded 'PRODUCT_CATEGORY' column based on mapping.")
    print("Mapping used:", category_to_product)
    print("\nFirst 5 rows with the new column:")
    # Display relevant columns including the new one
    print(df_customer[['ORDER_NUMBER', 'SERVICE_CATEGORY', 'PRODUCT_CATEGORY', 'ACCEPTANCE_TIME']].head().to_markdown(index=False))

    print("\nProduct categories for this customer:")
    print(df_customer['PRODUCT_CATEGORY'].value_counts())

elif 'df_customer' in locals() and df_customer.empty:
     print("No data for the selected customer to map.")
else:
    print("Error: df_customer DataFrame not found or is empty. Please ensure Step 4 ran successfully.")
```

```python
from langchain.chat_models import ChatOpenAI  # Import ChatOpenAI
from dotenv import dotenv_values  # Import dotenv_values

# Define your OpenAI API key
env_values = dotenv_values(".env")
open_api_key = env_values["open_api_key"]
# Use ChatOpenAI instead of OpenAI
llm = ChatOpenAI(openai_api_key=open_api_key, model_name="gpt-4o-mini")
```

```python
# Step 7: LLM Summarization (Using your initialized 'llm' object)

# (Keep the formatting function from the previous example)
def format_ticket_data_for_prompt(ticket_data_df):
    """Formats the DataFrame into a string suitable for an LLM prompt."""
    if ticket_data_df.empty:
        return "No ticket data available for this period."

    details = ["Ticket History:"]
    # Convert relevant columns to string and handle potential missing values (NaN)
    for index, row in ticket_data_df.iterrows():
        entry = (
            f"- Ticket: {row.get('ORDER_NUMBER', 'N/A')}, "
            f"Accepted: {row.get('ACCEPTANCE_TIME', 'N/A')}, "
            f"Completed: {row.get('COMPLETION_TIME', 'N/A')}\n"
            f"  Desc1: {row.get('ORDER_DESCRIPTION_1', 'N/A')}, "
            f"Desc2: {row.get('ORDER_DESCRIPTION_2', 'N/A')}, "
            f"Desc3: {row.get('ORDER_DESCRIPTION_3_MAXIMUM', 'N/A')}\n"
            f"  Note: {row.get('NOTE_MAXIMUM', 'N/A')}\n"
            f"  Completion Note: {row.get('COMPLETION_NOTE_MAXIMUM', 'N/A')}"
        )
        details.append(entry)
    return "\n".join(details)


# Modified function to use your 'llm' object (ChatOpenAI)
def get_llm_summary_with_chatopenai(llm_model, product_name, ticket_data_df):
    """Gets a summary using the provided ChatOpenAI llm object."""
    if llm_model is None:
        return "Error: LLM model object not provided."
    if ticket_data_df.empty:
        return f"No ticket data provided for {product_name} to summarize."

    # Format the input data (same as before)
    formatted_data = format_ticket_data_for_prompt(ticket_data_df)

    # Construct the detailed prompt (same as before) [cite: 17, 20, 23, 26, 29]
    prompt = f"""
    Please generate a detailed "storytelling" summary for a customer's experience with our '{product_name}' service, based ONLY on the following ticket history.
    Organize the summary into exactly these five sections: Initial Issue, Follow-ups, Developments, Later Incidents, and Recent Events. [cite: 17, 20, 23, 26, 29]
    For each section, determine the relevant timeframe (e.g., "early November 2024", "mid-November 2024") and list the corresponding ticket numbers (ORDER_NUMBER). [cite: 17, 20, 23, 26, 29]
    Focus on the narrative described in the descriptions and notes provided in the ticket history.

    Ticket History:
    {formatted_data}

    Generate the five-section summary now:
    """

    try:
        print(f"\n--- Calling LLM ({llm_model.model_name}) for product: {product_name} ---")

        # Make the API call using the .invoke() method common for LangChain/ChatOpenAI
        # (Ensure you have imported necessary classes like HumanMessage if using that pattern,
        # otherwise invoking with the string directly might work depending on your setup)
        # from langchain_core.messages import HumanMessage # Example import
        # response = llm_model.invoke([HumanMessage(content=prompt)]) # LangChain pattern
        response = llm_model.invoke(prompt) # More direct pattern if supported

        print(f"--- LLM ({llm_model.model_name}) call successful ---")

        # Extract the summary from the response object
        # For LangChain, the content is often in response.content
        summary = response.content if hasattr(response, 'content') else str(response)
        return summary.strip()

    except Exception as e:
        print(f"--- Error during LLM call: {e} ---")
        return f"Error generating summary for {product_name}: {e}"

# --- Using the function ---

# Check if 'summary_input_data' exists from Step 6 and your 'llm' object exists
# ('llm' should be the ChatOpenAI object you created before this cell)
if 'summary_input_data' in locals() and 'llm' in locals() and llm is not None:
    # Call the function with the prepared data for the example product
    # Make sure 'example_product' variable is set from Step 6
    print(f"Generating summary for '{example_product}' using {llm.model_name}...")
    actual_summary = get_llm_summary_with_chatopenai(llm, example_product, summary_input_data) # Pass your 'llm' object
    print("\n--- Generated Summary ---")
    print(actual_summary)
    print("--- End of Summary ---")

elif 'llm' not in locals() or llm is None:
     print("\nCannot generate summary: LLM object ('llm') not found or not initialized.")
elif 'summary_input_data' not in locals():
    print(f"\nCannot generate summary for '{example_product}': Input data not prepared in Step 6.")
else:
     print("\nCannot generate summary: Unknown reason.")
```
